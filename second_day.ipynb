```python
x = range(0, 5)
y1 = [2, 5, 7, 8, 10]
y2 = [3, 6, 8, 9, 11]

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
print(ax.lines)
```

```python
x = range(0, 5)
y1 = [2, 5, 7, 8, 10]
y2 = [3, 6, 8, 9, 10]
fig, ax = plt.subplots()
lines = [Line2D(x, y1), Line2D(x, y2, color='orange')]
for line in lines:
    ax.add_line(line)
ax.set_xlim(0, 4)
ax.set_ylim(2, 11)
```
